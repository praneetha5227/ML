```python
import numpy as np

def calculate_class_conditional_densities(feature_values, class_values):
    class_conditional_densities = {}

    for feature in feature_values:
        for class_value in np.unique(class_values['buys_computer']):
            # Calculate the probability density function (PDF) for the feature given the class
            feature_class_data = list(zip(feature_values[feature], class_values['buys_computer']))
            feature_class_values = [value[0] for value in feature_class_data if value[1] == class_value]
            
            # Use np.unique with return_counts=True
            feature_values_unique, feature_freq = np.unique(feature_class_values, return_counts=True)
            class_conditional_densities[(feature, class_value)] = {key: value / len(feature_class_values) for key, value in zip(feature_values_unique, feature_freq)}

    return class_conditional_densities

# Example data (replace this with your actual data)
data = {
    'age': ['<=30', '<=30', '31-40', '>40', '>40', '>40', '31-40', '<=30', '<=30', '>40', '<=30', '31-40', '31-40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}
feature_values = {
    'age': data['age'],
    'income': data['income'],
    'student': data['student'],
    'credit_rating': data['credit_rating'],
}

class_values = {
    'buys_computer': data['buys_computer'],
}

class_conditional_densities = calculate_class_conditional_densities(feature_values, class_values)

print("Class Conditional Densities:")
for key, value in class_conditional_densities.items():
    print(f"{key}: {value}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Given data
data = {
    'age': ['<=30', '<=30', '31-40', '>40', '>40', '>40', '31-40', '<=30', '<=30', '>40', '<=30', '31-40', '31-40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Select the columns for which you want to test independence
columns_to_test = ['age', 'income', 'student', 'credit_rating']

# Perform the chi-squared test for each feature
for feature in columns_to_test:
    contingency_table = pd.crosstab(index=df['buys_computer'], columns=df[feature])
    chi2, p, _, _ = chi2_contingency(contingency_table)

    # Output the results
    print(f"\nChi-squared statistic for {feature}: {chi2}")
    print(f"P-value: {p}")

    # Interpret the results
    alpha = 0.05  # Set your significance level
    if p < alpha:
        print("Reject the null hypothesis (There is a significant association between the feature and the target variable).")
    else:
        print("Fail to reject the null hypothesis (There is no significant association between the feature and the target variable).")

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, classification_report

# Given data
data = {
    'age': ['<=30', '<=30', '31-40', '>40', '>40', '>40', '31-40', '<=30', '<=30', '>40', '<=30', '31-40', '31-40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert categorical variables to numerical labels
df_encoded = pd.get_dummies(df.drop('buys_computer', axis=1))

# Extract target variable
target = df['buys_computer']

# Split the data into training and testing sets
Tr_X, Ts_X, Tr_y, Ts_y = train_test_split(df_encoded, target, test_size=0.2, random_state=42)

# Build and train the Naïve-Bayes classifier
model = CategoricalNB()
model.fit(Tr_X, Tr_y)

# Make predictions on the test set
predictions = model.predict(Ts_X)

# Evaluate the model
accuracy = accuracy_score(Ts_y, predictions)
classification_report_str = classification_report(Ts_y, predictions)

# Output the results
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_report_str)
```
